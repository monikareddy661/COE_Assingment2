```python
# Calculate total and average sales per month
monthly_sales=stores_sales[['Jan','Feb','Mar','Apr']]
# Total sales per month
total_sales = monthly_sales.sum()
print(total_sales)
```

```python
# Average sales per month
average_sales = monthly_sales.mean()
print( average_sales)
```

```python
# Total yearly sales for all stores combined
total_yearly_sales_all_stores = stores_sales[['Jan','Feb','Mar','Apr']].sum()
print(total_yearly_sales_all_stores)
```

```python
# Add Total_Sales column (sum of Jan to Apr)
stores_sales['TOTAL_SALES']=stores_sales[['Jan','Feb','Mar','Apr']].sum(axis=1)
stores_sales.head()
```

```python
# Top 5 stores with highest Total_Sales
top_5= stores_sales.sort_values(by='TOTAL_SALES',ascending=False).head(5)
top_5[['store_id','TOTAL_SALES']]
```

```python
# Bottom 5 stores with lowest Total_Sales
bottom_5=stores_sales.sort_values(by='TOTAL_SALES',ascending=True).head(5)
bottom_5[['store_id','TOTAL_SALES']]
```

```python
# Find store with maximum sales in March
max_march=stores_sales.loc[stores_sales['Mar'].idxmax(), 'store_id']
max_march
```

```python
# Find store with maximum sales in April
min_apr=stores_sales.loc[stores_sales['Apr'].idxmin(), 'store_id']
min_apr
```

```python
# Filter stores from California that had sales > 20 in any month.
stores_sales[(stores_sales['city']=='California')&((stores_sales['Jan']>20)|(stores_sales['Feb']>20)|(stores_sales['Mar']>20)|(stores_sales['Apr']>20))][['store_id','Jan','Feb','Mar','Apr','TOTAL_SALES']]
```

```python
# Calculate average sales per store across all months (Jan–Apr)
stores_sales['AVERAGE_SALES']=(stores_sales['Jan']+stores_sales['Feb']+stores_sales['Mar']+stores_sales['Apr']) / 4
stores_sales.head()
```

```python
# Sort stores by average sales in descending order
stores_sales_sorted = stores_sales.sort_values(by='AVERAGE_SALES', ascending=False)
stores_sales_sorted[['store_id','Jan','Feb','Mar','Apr','AVERAGE_SALES']]
```

```python
# Calculate total sales for each month
total_per_month=stores_sales[['Jan','Feb','Mar','Apr']].sum()
# Find the month with highest total sales
month_highest=total_per_month.idxmax()
month_highest
```

```python
# For each store, find the month with maximum sales
stores_sales['MONTH_MAX'] = stores_sales[['Jan','Feb','Mar','Apr']].idxmax(axis=1)
stores_sales[['store_id','Jan','Feb','Mar','Apr','MONTH_MAX']]
```
